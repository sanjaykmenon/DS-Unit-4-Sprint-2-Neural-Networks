## Write your own perceptron code that can correctly classify (99.0% accuracy) a NAND gate. 

| x1 | x2 | y |
|----|----|---|
| 0  | 0  | 1 |
| 1  | 0  | 1 |
| 0  | 1  | 1 |
| 1  | 1  | 0 |

```python
import pandas as pd
data = { 'x1': [0,1,0,1],
         'x2': [0,0,1,1],
         'y':  [1,1,1,0],
         'b': [1,1,1,1]
       }

df = pd.DataFrame.from_dict(data).astype('int')
```

```python
##### Your Code Here #####
df.shape
```

```python
inputs = df
```

```python
outputs = ([1],[1],[1],[0])
```

```python
import numpy as np
```

```python
weights = 2 * np.random.random((4,1)) - 1
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
```

```python
weighted_sum = np.dot(inputs, weights)
```

```python
# first output
activated_output = sigmoid(weighted_sum)
activated_output
```

```python
# Find the error
error = data['y'] - activated_output
error
```

```python
# Gradient descent, finding the steepest slope to lower error

adjustments = error * sigmoid_derivative(weighted_sum)
adjustments.shape
```

```python
# Forward Propagation time

for i in range(10000):
  # dot product the inputs and weights to get weighted sum
  weighted_sum = np.dot(inputs, weights)
  
  # activation function, could add bias here if you wanted
  activated_output = sigmoid(weighted_sum)

  # actual - predicted = error
  error = outputs - activated_output

  # Back Prop time
  adjustments = error * sigmoid_derivative(weighted_sum)

  # Update wthe weights with the dot product of inputs.T and adjustments
  weights = weights + np.dot(inputs.T, adjustments)
print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)
```

```python
diabetes = pd.read_csv('https://raw.githubusercontent.com/ryanleeallred/datasets/master/diabetes.csv')
diabetes.head()
```

Although neural networks can handle non-normalized data, scaling or normalizing your data will improve your neural network's learning speed. Try to apply the sklearn `MinMaxScaler` or `Normalizer` to your diabetes dataset. 

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer

features = list(diabetes)[:-1]
target = list(diabetes)[-1:]

#instantiate normalizer. This is used because different features have different ranges.

normalizer = Normalizer()

#normalize feature data

X = normalizer.fit_transform(diabetes[features])

#target
y = diabetes[target]

```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
##### Update this Class #####

class Perceptron:
    
    def __init__(self, niter = 10):
        self.niter = niter
    
    def __sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def __sigmoid_derivative(self, x):
        sx = sigmoid(x)
        return sx * (1-sx)

    def fit(self, X, y):  
       # Randomly Initialize Weights
        self.weight = 2 * np.random.random((8,1)) - 1 #8 here is the number of input vectors. Outcome column is our target / prediction variable

        for i in range(self.niter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(X, self.weight) #input vectors multiplied by randomly assigned weights.
            # Activate!
            activated_output = self.__sigmoid(weighted_sum) #weighted sum is passed through an activation function, here it is a sigmoid function.
            # Cac error
            error = y - activated_output #checking for error by difference between activated output and the target y 
            adjustments = error * self.__sigmoid_derivative(weighted_sum) #adjustment value that is used to redefine weights in the next step
            # Update the Weights
            self.weight = self.weight + np.dot(X.T, adjustments) #adding the adjustments to existing randomized weight values to make it closer to target y.


    def predict(self, X):
        """Return class label after unit step"""
        weighted_sum = np.dot(X, self.weight) + 1
        activated_output = self.__sigmoid(weighted_sum)
        return np.round(activated_output)
```

```python
model = Perceptron()
model.fit(X,y)
model.predict(X)
```
